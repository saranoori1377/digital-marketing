```python
import os
import json
import pandas as pd


# مشخص کردن نام فایل‌های JSON
pc_data = ['apple.json', 'acer.json', 'asus.json', 'hp.json', 'lenovo.json', 'msi.json']
mobile_data = ['honor.json', 'huawei.json', 'nokia.json', 'samsung.json', 'xiaomi.json']

all_pc_data = []
all_mobile_data = []

# خواندن داده‌های کامپیوتر
for filename in os.listdir():
    if filename in pc_data:  
        with open(filename, 'r', encoding='utf-8') as file:
            data1 = json.load(file)
            all_pc_data.append(data1) if isinstance(data1, dict) else all_pc_data.extend(data1)

# خواندن داده‌های موبایل
for filename in os.listdir():
    if filename in mobile_data:  
        with open(filename, 'r', encoding='utf-8') as file:
            data2 = json.load(file)
            all_mobile_data.append(data2) if isinstance(data2, dict) else all_mobile_data.extend(data2)

# تبدیل داده‌ها به DataFrame
df_mobile = pd.DataFrame(all_mobile_data)
df_pc = pd.DataFrame(all_pc_data)

# تعیین نام ستون‌ها برای موبایل و کامپیوتر
columns1 = ['id', 'link', 'name', 'price', 'rate', 'free send', 'status']
columns2= ['id','link','image','name','price','rate','freesend', 'status']
if not df_mobile.empty:
    df_mobile.columns1 = columns1
if not df_pc.empty:
    df_pc.columns2 = columns2
```

```python
import pandas as pd

# تابع تبدیل اعداد فارسی به انگلیسی
def convert_farsi_to_english_numbers(s):
    """تبدیل اعداد فارسی به اعداد انگلیسی"""
    numbers = {
        '۰': '0', '۱': '1', '۲': '2', '۳': '3', '۴': '4',
        '۵': '5', '۶': '6', '۷': '7', '۸': '8', '۹': '9'
    }
    if isinstance(s, str):
        for persian_digit, english_digit in numbers.items():
            s = s.replace(persian_digit, english_digit)
    return s

# پردازش داده‌های قیمت‌ها و نرخ‌ها
def process_row(row):
    """پردازش یک سطر از داده‌ها"""
    # تبدیل اعداد فارسی
    row['price'] = convert_farsi_to_english_numbers(row['price'])
    row['rate'] = convert_farsi_to_english_numbers(row['rate'])
    
    # حذف درصد و کاما از قیمت و نرخ
    row['price'] = str(row['price']).replace('٪', '').replace(',', '')
    row['rate'] = str(row['rate']).replace('٪', '').replace(',', '')

    # تبدیل به نوع عددی
    row['price'] = pd.to_numeric(row['price'], errors='coerce')
    row['rate'] = pd.to_numeric(row['rate'], errors='coerce')
    
    return row

# حلقه برای پردازش سطرها
def process_dataframe(df):
    processed_rows = []
    for idx, row in df.iterrows():
        # اگر سطر شامل NaN نباشد، پردازش شود
        if not row.isnull().any():
            processed_row = process_row(row)
            processed_rows.append(processed_row)
    # ساخت DataFrame جدید از سطرهای پردازش شده
    return pd.DataFrame(processed_rows)

# مثال استفاده برای df_pc و df_mobile
if df_pc is not None:
    df_pc_processed = process_dataframe(df_pc)
    print("PC Prices and Rates after processing:")
    print(df_pc_processed[['price', 'rate']].head(5))

if df_mobile is not None:
    df_mobile_processed = process_dataframe(df_mobile)
    print("Mobile Prices and Rates after processing:")
    print(df_mobile_processed[['price', 'rate']].head(5))
```

```python
import pandas as pd

# تابع تبدیل اعداد فارسی به انگلیسی
def convert_farsi_to_english_numbers(s):
    """تبدیل اعداد فارسی به اعداد انگلیسی"""
    numbers = {
        '۰': '0', '۱': '1', '۲': '2', '۳': '3', '۴': '4',
        '۵': '5', '۶': '6', '۷': '7', '۸': '8', '۹': '9'
    }
    if isinstance(s, str):
        for persian_digit, english_digit in numbers.items():
            s = s.replace(persian_digit, english_digit)
    return s

# پردازش داده‌های قیمت‌ها و نرخ‌ها
def process_row(row):
    """پردازش یک سطر از داده‌ها"""
    # تبدیل اعداد فارسی
    row['price'] = convert_farsi_to_english_numbers(row['price'])
    row['rate'] = convert_farsi_to_english_numbers(row['rate'])
    
    # حذف درصد و کاما از قیمت و نرخ
    row['price'] = str(row['price']).replace('٪', '').replace(',', '')
    row['rate'] = str(row['rate']).replace('٪', '').replace(',', '')

    # تبدیل به نوع عددی
    row['price'] = pd.to_numeric(row['price'], errors='coerce')
    row['rate'] = pd.to_numeric(row['rate'], errors='coerce')
    
    return row

# تابعی برای پردازش DataFrame و تغییر ساختار آن
def transform_dataframe(df):
    """تبدیل ساختار DataFrame و پردازش سطرها"""
    processed_rows = []
    
    for idx, row in df.iterrows():
        # پردازش سطر
        processed_row = process_row(row)
        processed_rows.append(processed_row)

    # ساخت DataFrame جدید از سطرهای پردازش شده
    transformed_df = pd.DataFrame(processed_rows)

    # حذف سطرهای خالی
    transformed_df = transformed_df.dropna(subset=['price', 'rate'])

    # مرتب‌سازی DataFrame بر اساس قیمت
    transformed_df = transformed_df.sort_values(by='price').reset_index(drop=True)

    return transformed_df

# مثال استفاده برای df_pc و df_mobile
if df_pc is not None and not df_pc.empty:
    df_pc_transformed = transform_dataframe(df_pc)
    print("Transformed PC Data:")
    print(df_pc_transformed[['name', 'price', 'rate']].head(5))

if df_mobile is not None and not df_mobile.empty:
    df_mobile_transformed = transform_dataframe(df_mobile)
    print("Transformed Mobile Data:")
    print(df_mobile_transformed[['name', 'price', 'rate']].head(5))
```

```python
import pandas as pd
import numpy as np

# فرض کنید df_pc و df_mobile از قبل تعریف شده‌اند

# تابعی برای تبدیل اعداد فارسی به انگلیسی
def convert_farsi_to_english_numbers(s):
    """تبدیل اعداد فارسی به اعداد انگلیسی"""
    numbers = {
        '۰': '0', '۱': '1', '۲': '2', '۳': '3', '۴': '4',
        '۵': '5', '۶': '6', '۷': '7', '۸': '8', '۹': '9'
    }
    if isinstance(s, str):
        for persian_digit, english_digit in numbers.items():
            s = s.replace(persian_digit, english_digit)
    return s

# تابعی برای پردازش قیمت
def process_price(price):
    """پردازش قیمت برای حذف علایم و تبدیل به عدد"""
    # تبدیل اعداد فارسی
    price = convert_farsi_to_english_numbers(price)
    # حذف درصد و کاما
    price = str(price).replace('٪', '').replace(',', '').strip()
    # تبدیل به عدد
    return pd.to_numeric(price, errors='coerce')

# تابعی برای پر کردن مقادیر گمشده با استفاده از نام
def fill_missing_values(df):
    # اطمینان از اینکه همه مقادیر در ستون name از نوع رشته هستند
    df['name'] = df['name'].astype(str)

    # تابعی برای پر کردن مقادیر گمشده
    def fill_row(row):
        if pd.isnull(row['price']):
            # پیدا کردن دستگاه‌های مشابه بر اساس نام
            similar_items = df[df['name'].str.contains(row['name'], na=False, case=False)]
            if not similar_items.empty:
                # پر کردن با قیمت متوسط دستگاه‌های مشابه
                row['price'] = similar_items['price'].mean()

        if pd.isnull(row['rate']):
            # پیدا کردن دستگاه‌های مشابه بر اساس نام
            similar_items = df[df['name'].str.contains(row['name'], na=False, case=False)]
            if not similar_items.empty:
                # پر کردن با امتیاز متوسط دستگاه‌های مشابه
                row['rate'] = similar_items['rate'].mean()

        # پردازش قیمت و نرخ
        row['price'] = process_price(row['price'])
        row['rate'] = pd.to_numeric(row['rate'], errors='coerce')  # تبدیل نرخ به عدد

        return row

    # استفاده از apply برای پر کردن سطرها
    df = df.apply(fill_row, axis=1)
    return df

# پر کردن مقادیر گمشده برای کامپیوتر
if df_pc is not None and not df_pc.empty:
    df_pc = fill_missing_values(df_pc)
    print("PC Data after filling missing values:")
    print(df_pc[['name', 'price', 'rate']].head(5))
else:
    print("Cannot fill missing values for df_pc because it is None or empty.")

# پر کردن مقادیر گمشده برای موبایل
if df_mobile is not None and not df_mobile.empty:
    df_mobile = fill_missing_values(df_mobile)
    print("\nMobile Data after filling missing values:")
    print(df_mobile[['name', 'price', 'rate']].head(5))
else:
    print("Cannot fill missing values for df_mobile because it is None or empty.")
```

```python
import pandas as pd
import re

def normalize_persian_text(text):
    text = re.sub(r'\s+', ' ', text)  # حذف فاصله‌های اضافی
    text = text.replace('\u200c', ' ')  # جایگزینی نیم‌فاصله با فاصله معمولی
    return text.strip()

def extract_brand(name):
    brands = {'آنر', 'هوآوی', 'سامسونگ', 'شیائومی', 'اپل','نوکیا', 'ایسوس', 
              'اچ‌پی', 'اچ پی', 'لنوو', 'ام‌اس‌آی', 'ام اس آی', 'ایسر'}
    
    normalized_name = normalize_persian_text(name)  # استانداردسازی متن ورودی
    parts = normalized_name.split(' ')
    for part in parts:
        if part in brands:
            return part
    return None  # در صورت عدم یافتن برند

def add_brand_column(df, category_name):
    if not df.empty:
        # اگر process_columns وجود دارد، آن را فراخوانی کنید
        # df = process_columns(df)

        # اطمینان از اینکه مقادیر به رشته تبدیل شوند
        df['name'] = df['name'].astype(str).fillna('')  # استانداردسازی ستون 'name'
        df['brand'] = df['name'].apply(extract_brand)
        print(f"brand for{category_name}:")
        print(df['brand'].head())
    else:
        print(f"there is no data for : {category_name}.")

# فرض کنید df_mobile و df_pc از قبل تعریف شده‌اند
# استفاده از تابع برای هر دو دیتافریم موبایل و کامپیوتر
add_brand_column(df_mobile, " mobile")
add_brand_column(df_pc, " pc")
```

```python
# تابع استخراج مدل
def extract_model(name):
    # یافتن کلمه "مدل" و کلمات بعد از آن
    match = re.search(r'مدل (.+)', name)
    if match:
        return match.group(1).strip()  # کلمات بعد از "مدل"
    return None

# ایجاد ستون جدید برای مدل
df_mobile['model'] = df_mobile['name'].apply(extract_model)
df_pc['model'] = df_pc['name'].apply(extract_model)
```
